```python
import matplotlib.pyplot as plt
import numpy as np
H_in_kJmol = 2625.499639
```

```python
mass1 = np.loadtxt('mass/m1.out', skiprows=1)
mass5 = np.loadtxt('mass/m5.out', skiprows=1)
mass10 = np.loadtxt('mass/m10.out', skiprows=1)
mass20 = np.loadtxt('mass/m20.out', skiprows=1)
mass30 = np.loadtxt('mass/m30.out', skiprows=1)
mass50 = np.loadtxt('mass/m50.out', skiprows=1)

mass1[:,3] *= H_in_kJmol
mass5[:,3] *= H_in_kJmol
mass10[:,3] *= H_in_kJmol
mass20[:,3] *= H_in_kJmol
mass30[:,3] *= H_in_kJmol
mass50[:,3] *= H_in_kJmol

mass1[:,3] -= mass1[:,3].min()
mass5[:,3] -= mass5[:,3].min()
mass10[:,3] -= mass10[:,3].min()
mass20[:,3] -= mass20[:,3].min()
mass30[:,3] -= mass30[:,3].min()
mass50[:,3] -= mass50[:,3].min()

x = mass1[:,0]
```

```python
plt.rcParams["figure.figsize"] = [10,7]
fig, axs = plt.subplots(ncols=1)
axs.plot(x, mass1[:,3], label='m=1 a.u.')
axs.plot(x, mass5[:,3], label='m=5 a.u.')
axs.plot(x, mass10[:,3], label='m=10 a.u.')
#axs.plot(x, mass20[:,3], label='m=20 a.u.')
axs.plot(x, mass30[:,3], label='m=30 a.u.')
axs.plot(x, mass50[:,3], label='m=50 a.u.')

axs.grid()
axs.set_xlabel('torsion [degree]')
axs.set_ylabel('free Energy [Hartree]')
axs.set_title('free Energy')
plt.tight_layout()
plt.legend()
plt.savefig("mass.png", dpi=400)
plt.show()
plt.close()
```
